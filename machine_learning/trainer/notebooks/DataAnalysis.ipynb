```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import joblib


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("../data/mock_data.csv")
df.describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.bar(['Lost (0)', 'Won (1)'], [sum(df['user_won'] == 0), sum(df['user_won'] == 1)], color=['salmon', 'steelblue'])
plt.title('Distribution of Target Variable (user_won)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 5, figsize=(18, 4))

columns = ['minutes_slept', 'minutes_awake', 'env_score', 'light', 'target']

for ax, col in zip(axes, columns):
    if col == 'target':
        ax.bar(['Lost (0)', 'Won (1)'], df[col].value_counts().sort_index(), color=['salmon', 'steelblue'])
    else:
        ax.hist(df[col], bins=30, color='steelblue', edgecolor='white')
    ax.set_title(col)
    ax.set_ylabel('Count')

plt.suptitle('Feature Distributions', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```
